# 2. GRAFICOS PARA VISUALIZAÇÃO

#PLOTAR OS DADOS = CRIAR UM GRAFICO

* x = Informar a coluna do eixo X
* y = Informar a coluna do eixo y
* kind='bar'  *define o tipo de gráfico como de barras.*
* figsize=(10,7)  * define o tamanho da figura do gráfico*
* yaxis *se refere ao eixo y*.
* set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))    =  *define que os valores no eixo y serão formatados como números de ponto flutuante com duas casas decimais e separador de milhares.*

* plt.xticks(rotation=45)    *Inclinar as legendas em 45 graus*
* plt.title("TITULO"): *Adiciona um título ao gráfico.*
* plt.show():   *Exibe o gráfico na tela*


```python
barras_ago2008 = dados.plot(x = "Unidade da Federação", y = "2008/Ago", kind = "bar", figsize=(10,7))
barras_ago2008.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

linhas_ago2008 = dados.plot(x = "Unidade da Federação", y = "2008/Ago", figsize=(10,7))
linhas_ago2008.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

# Adicione a legenda
plt.title("VALOR UNIDADE")

# Exiba o gráfico
plt.show(barras_ago2008)
plt.show(linhas_ago2008)
```

```python
dados_usaveis["Total"] = dados_usaveis.sum(axis=1)

#Ordenando forma decrecente de valot Total
dados_usaveis = dados_usaveis.sort_values(by="Total", ascending=False)
dados_usaveis.head()
```
