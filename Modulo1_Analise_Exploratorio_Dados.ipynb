#IMPORTANDO A BIBLIOTECA PANDAS

```python
import pandas as pd
```

#CARREGANDO A TABELA E TRATANDO AS COLUNAS
encoding='ISO-8859-1 = Transformando em UTF-8
skiprows=3  = Desconsidera a 3 primeiras linhas da tabela
sep = ";"   = Informando que o separador é ;
skipfooter=12   = Excluindo as 12 linhas finais que tem informações que não são dados
Thousands = "."   = Trocando a , por ponto em separador de milhar no valor
decimal = ","   = Informando que o separador de casas decimais é ,

```python
dados = pd.read_csv('/content/A152918189_28_143_208.csv', encoding='ISO-8859-1',
            skiprows=3, sep = ";" , skipfooter=12, thousands=".", decimal=",")
print(dados)
```
