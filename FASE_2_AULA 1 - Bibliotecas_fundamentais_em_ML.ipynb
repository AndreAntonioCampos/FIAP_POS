CRIAR NOVOS VALORES E ARMAZENAR DENTRO DAS COLUNAS

Fazer uma nova variavel, criando uma função

Criar uma função def para categorizar os tomates e depois incluir essa informação na minha base de dados

```python
df_csv.head()
```

```python
#def = função da variavel
#categorizar_tomate_media = nome da minha função
#(media) a coluna
#if = se
#elis = um segundo se
#else = Se não achar o if ou elif
#return = Valor que vai retornar na minha variavel

def categorizar_tomate_media(media):
  if media >=40 and media <=70:
    return "tomate medio"
  elif media <40:
    return "tomate pequeno"
  else:
    return "tomatão"
```

```python
#CRIAR A NOVA COLUNA
#df["coluna que vou criar"] = df["por qual coluna faz a leitura para aplicar"].apply(a função que criei)
df_csv["categoria_tomate"] = df_csv["Average"].apply(categorizar_tomate_media)
```

```python
df_csv.head()
```

```python
#Para ver a média de acordo com uma coluna usando o describe()
#groupby = agrupar por
#(["categoria_tomate"]) = informando o nome da coluna
#describe() = função para ver a média

df_csv.groupby(["categoria_tomate"]).describe()
```

Array

OPERAÇÕES MATEMATICAS

```python
#Criando lista dentro de lista porque é possivel separar valores
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
arr3 = np.array([[9,10],[11,12]])
```

```python
#Array para concatenar as anteriores np.concatenate, o axis=1 estou informando qual o eixo da matriz
arr4 = np.concatenate((arr1,arr2,arr3),axis=1)
print(arr4)
```

```python
#Para dividir os Arrays (split), USEI O 2 PORQUE EU QUERO SEPARAR EM 2 ARRAYS
arr4_split = np.split(arr4,2)
print(arr4_split)
```

```python
#Transpor para facilitar calculos (np.transpose)
arr4_transpose = np.transpose(arr4)
print(arr4_transpose)
```

```python
#Se eu quiser reverter para não ter que rodar todo o código
arr4_revertido = arr4_transpose.T
print(arr4_revertido)
```

```python
#Operações Matemáticas add (adição)
arr_a = np.array([1,7,27])
arr_b = np.array([5,8,30])
arr_a_b = np.add(arr_a, arr_b)
print(arr_a_b)
```

```python
#Operações Matemáticas subtract (subtração)
arr_sub_a_b = np.subtract(arr_a, arr_b)
print(arr_sub_a_b)
```

```python
#Operações Matemáticas multiply (Multiplicação)
arr_multiplica_a_b = np.multiply(arr_a, arr_b)
print(arr_multiplica_a_b)
```

```python
#Operações Matemáticas divide (divisão)
arr_divide_a_b = np.divide(arr_a, arr_b)
print(arr_divide_a_b)
```

```python
#Operações Matemáticas mean (média)
arr_media_a_b = np.mean(arr_a)
print(arr_media_a_b)
```

```python
#Operações Matemáticas percentil (Percentual, neste caso por 78%)
arr_percentual_a_b = np.percentile(arr_a,78)
print(arr_percentual_a_b)
```

PRECISAMOS SEPARAR A BASE DE DADOS EM TESTE PARA QUE O ALGORITIMO RECONHEÇA A BASE, ELE TREINE

POR VOLTA DE 70% A 80% DA BASE DE TREINO
POR VOLTA DE 30% A 20% DA BASE DE TESTE

Para que o algoritimo aprenda os padrões

```python
#Import de uma parte da biblioteca
from sklearn.model_selection import train_test_split
```

```python
#Separar os dados na variavel X e variavel Y
#Variavel caracteristica na Variavel X. Todas as minhas colunas ou colunas especificas
#Variavel target na Variavel Y

#Estou excluindo a Class variable do eixo X
x = df_diabetes.drop(["Class variable"], axis=1)

#Estou incluindo a Class variable no eixo Y
y = df_diabetes["Class variable"]
```

```python
y
```

**FASE 2 - AULA 1 - EPISÓDIO 9  - BIBLIOTECAS FUNDAMENTAIS E PRIMEIROS PASSOS EM ML**

Para modelos de machine Learning a principal é:

**  scikit-learn **


```python
#Armazenar os dados na nossa base de treino e uma na de teste
#x_train = Base de treino
#x_test = resposta do x
#y_train = Base de treino
#y_test = resposta do y
#train_test_split() = Onde vou fazer a separação dos dados
#x,y = Qual base ele vai olhar
#test_size = 0.3 = Tamanho de 30% pro ambiente de teste

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
#Essa é a base de treinamento
#Separarou 537 linhas de maneira aleatória
x_train
```

```python
#Essa é a base de treinamento
#Separarou 231 linhas de maneira aleatória
x_test
```

**ESCOLHER ALGUM MODELO DE MACHINE LEARNING (Algoritimo)**

Vamos usar de classificação por caracteristica da base. Usaremos o kneighbors  o knn(numero de vizinhos)

```python
#Importar o kneighbors
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Passar os paramêtros
#n_neighbors=3) = estou dizendo pra ele olhar os 3 vizinhos mais próximos
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
#Treinar o algoritimo
#fit = treinar conhecendo os dados
#x_train, y_train = tanto os dados quanto a resposta
knn.fit(x_train, y_train)
```

```python
# Testar para entender se está efetivo ou não, O quanto nosso modelo acertou
#x_test e Y_test = A base de treino
accuracy = knn.score(x_test, y_test)
```
