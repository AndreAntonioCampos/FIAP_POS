CRIAR NOVOS VALORES E ARMAZENAR DENTRO DAS COLUNAS

Fazer uma nova variavel, criando uma fun√ß√£o

Criar uma fun√ß√£o def para categorizar os tomates e depois incluir essa informa√ß√£o na minha base de dados

```python
df_csv.head()
```

```python
#def = fun√ß√£o da variavel
#categorizar_tomate_media = nome da minha fun√ß√£o
#(media) a coluna
#if = se
#elis = um segundo se
#else = Se n√£o achar o if ou elif
#return = Valor que vai retornar na minha variavel

def categorizar_tomate_media(media):
  if media >=40 and media <=70:
    return "tomate medio"
  elif media <40:
    return "tomate pequeno"
  else:
    return "tomat√£o"
```

```python
#CRIAR A NOVA COLUNA
#df["coluna que vou criar"] = df["por qual coluna faz a leitura para aplicar"].apply(a fun√ß√£o que criei)
df_csv["categoria_tomate"] = df_csv["Average"].apply(categorizar_tomate_media)
```

```python
df_csv.head()
```

```python
#Para ver a m√©dia de acordo com uma coluna usando o describe()
#groupby = agrupar por
#(["categoria_tomate"]) = informando o nome da coluna
#describe() = fun√ß√£o para ver a m√©dia

df_csv.groupby(["categoria_tomate"]).describe()
```

**FASE 2 - AULA 4 - BIBLIOTECAS FUNDAMENTAIS E PRIMEIROS PASSOS EM ML**

Array

OPERA√á√ïES MATEMATICAS

```python
#Criando lista dentro de lista porque √© possivel separar valores
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
arr3 = np.array([[9,10],[11,12]])
```

```python
#Array para concatenar as anteriores np.concatenate, o axis=1 estou informando qual o eixo da matriz
arr4 = np.concatenate((arr1,arr2,arr3),axis=1)
print(arr4)
```

```python
#Para dividir os Arrays (split), USEI O 2 PORQUE EU QUERO SEPARAR EM 2 ARRAYS
arr4_split = np.split(arr4,2)
print(arr4_split)
```

```python
#Transpor para facilitar calculos (np.transpose)
arr4_transpose = np.transpose(arr4)
print(arr4_transpose)
```

```python
#Se eu quiser reverter para n√£o ter que rodar todo o c√≥digo
arr4_revertido = arr4_transpose.T
print(arr4_revertido)
```

```python
#Opera√ß√µes Matem√°ticas add (adi√ß√£o)
arr_a = np.array([1,7,27])
arr_b = np.array([5,8,30])
arr_a_b = np.add(arr_a, arr_b)
print(arr_a_b)
```

```python
#Opera√ß√µes Matem√°ticas subtract (subtra√ß√£o)
arr_sub_a_b = np.subtract(arr_a, arr_b)
print(arr_sub_a_b)
```

```python
#Opera√ß√µes Matem√°ticas multiply (Multiplica√ß√£o)
arr_multiplica_a_b = np.multiply(arr_a, arr_b)
print(arr_multiplica_a_b)
```

```python
#Opera√ß√µes Matem√°ticas divide (divis√£o)
arr_divide_a_b = np.divide(arr_a, arr_b)
print(arr_divide_a_b)
```

```python
#Opera√ß√µes Matem√°ticas mean (m√©dia)
arr_media_a_b = np.mean(arr_a)
print(arr_media_a_b)
```

```python
#Opera√ß√µes Matem√°ticas percentil (Percentual, neste caso por 78%)
arr_percentual_a_b = np.percentile(arr_a,78)
print(arr_percentual_a_b)
```
