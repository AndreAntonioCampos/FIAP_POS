#CRIANDO VARIAVEIS NUMERICAS PARA TRABALHAR COM MÉDIA DE VALORES

```python

# Selecionando apenas as colunas numéricas para trabalhar com médias de valores
dados_numericos = dados.select_dtypes(include='number')

# Preenchendo os valores vazios com 0
dados_vazios_preenchidos = dados_numericos.fillna(0)

# Adicionando a coluna "Unidade da Federação" de volta ao DataFrame
dados_vazios_preenchidos.insert(0, "Unidade da Federação", arquivo["Unidade da Federação"])

# A variável final é chamada "dados"
dados = dados_vazios_preenchidos

display(dados_numericos)
display(dados_vazios_preenchidos)

dados.head()
```

# 2. GRAFICOS PARA VISUALIZAÇÃO

#PLOTAR OS DADOS = CRIAR UM GRAFICO

* x = Informar a coluna do eixo X
* y = Informar a coluna do eixo y
* kind='bar'  *define o tipo de gráfico como de barras.*
* figsize=(10,7)  * define o tamanho da figura do gráfico*
* yaxis *se refere ao eixo y*.
* set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))    =  *define que os valores no eixo y serão formatados como números de ponto flutuante com duas casas decimais e separador de milhares.*

* plt.xticks(rotation=45)    *Inclinar as legendas em 45 graus*
* plt.title("TITULO"): *Adiciona um título ao gráfico.*
* plt.show():   *Exibe o gráfico na tela*


```python
barras_ago2008 = dados.plot(x = "Unidade da Federação", y = "2008/Ago", kind = "bar", figsize=(10,7))
barras_ago2008.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

linhas_ago2008 = dados.plot(x = "Unidade da Federação", y = "2008/Ago", figsize=(10,7))
linhas_ago2008.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

# Adicione a legenda
plt.title("VALOR UNIDADE")

# Exiba o gráfico
plt.show(barras_ago2008)
plt.show(linhas_ago2008)
```

# ********************************************************************************

```python
# Desafio 1: executar o gráfico anterior com o mês mais recente

#Com esse comando consigo visualizar no meu df a coluna mais recente que neste caso é 2021/Ago  dados.min() ou dados.max()
dados.min()

#Criando e formatando o Grafico
axis = dados.plot(x = 'Unidade da Federação',y='2021/Ago', figsize=(10,7), kind= 'bar')
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

# Adicionando a legenda
plt.title("AGO-2021")

# Exiba o gráfico
plt.show()
```

```python
#Desafio 2: deixar as legendas ânguladas (0, 45 etc) para facilitar a leitura

# Criando e formatando o gráfico
axis = dados.plot(x='Unidade da Federação', y='2021/Ago', figsize=(10, 7))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

# Adicionando a legenda
plt.title("AGO-2021")

# Inclinar as legendas em 45 graus
plt.xticks(rotation=45)

# Exiba o gráfico
plt.show()

```

#TRANSFORMANDO ESSE INDICE EM UMA LISTA

Com isso consigo usar apenas as colunas que tem dados

* df.tolist()
* df.insert()
* df[:5]

```python
colunas_usaveis = dados_numericos.mean().index.tolist()
#Como não queremos apenas as colunas de valores(Usaveis) vamos inserir a primeira coluna com os dados da Unidade da Federeção
#que no Python é a posição 0
colunas_usaveis.insert(0, "Unidade da Federação")

# Usando o colchetes[] com : numero, informamos quantos itens da lista de Python queremos imprimir. Não chamo o head porque é uma lista
colunas_usaveis[:5]
```
